```python
import  numpy as np
import  pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[4,5,9]

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title("test graph")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title("test graph")
plt.legend(['this is y','this is z'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
